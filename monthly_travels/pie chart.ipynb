```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/hallemorel/Downloads/monthly_travels/monthly_travel1819.csv')
```

```python
#remove all comas from numbers
cols = ['Nov-18', 'Dec-18','Jan-19', 'Feb-19', 'Mar-19', 'Apr-19', 'May-19', 'Jun-19', 'Jul-19', 'Aug-19', 'Sep-19', 'Oct-19', 'YTD']
# pass them to df.replace(), specifying each char and it's replacement:
df[cols] = df[cols].replace({',': '', ',': ''}, regex=True)
df
```

```python
#drop NaN row
df.drop(df.index[8])
```

```python
#create total number of tourists per region in a given year database
region_df = pd.DataFrame ({
    'Region': ['Europe', 'Caribbean', 'Asia', 'South America', 'Central America', 'Oceania', 'Middle East', 'Africa'],
    'YTD': [16611203, 7851695, 5356404, 1872581, 2814471, 758648, 2082536, 440550]
})
region_df
```

```python
#convert YTD column to numeric
region_df['YTD'] = pd.to_numeric(region_df['YTD'])
region_df
```

```python
#find the sum of the YTD column
region_df['YTD'].sum()
```

```python
#find the percent in which each region was traveled to over the year
percent = region_df['YTD']/37788088
region_df['Percent of Total']= percent

region_df
```

```python
labels = ["Europe", "Caribbean", "Asia", "South America", "Central America", "Oceania", "Middle East", "Africa"]
sizes = [44, 21, 15, 5, 7, 3, 5, 1]
explode = (0.1, 0, 0, 0, 0, 0, 0, 0)
```

```python
pie=plt.pie(sizes, explode=explode, labels=labels, autopct="%1.1f%%", startangle=200)
plt.title("Most Traveled to Regions in 2019")
plt.axis('equal')
```
