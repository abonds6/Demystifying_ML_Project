```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/hallemorel/Downloads/FinalProject/monthly_travel1819.csv')
```

```python
#remove all comas from numbers
cols = ['Nov-18', 'Dec-18','Jan-19', 'Feb-19', 'Mar-19', 'Apr-19', 'May-19', 'Jun-19', 'Jul-19', 'Aug-19', 'Sep-19', 'Oct-19', 'YTD']
# pass them to df.replace(), specifying each char and it's replacement:
df[cols] = df[cols].replace({',': '', ',': ''}, regex=True)
df
```

```python
#drop NaN row
df.drop(df.index[8])
```

```python
#create data base on the most traveled to region each month
top_df = pd.DataFrame({
    'Months': ['Nov-18', 'Dec-18','Jan-19', 'Feb-19', 'Mar-19', 'Apr-19', 'May-19', 'Jun-19', 'Jul-19', 'Aug-19', 'Sep-19', 'Oct-19'],
    'Region': ['Europe', 'Europe', 'Europe', 'Caribbean', 'Europe', 'Europe', 'Europe', 'Europe', 'Europe', 'Europe', 'Europe', 'Europe'],
    'Tourists': [1060561, 1204379, 825706, 791219, 1352526, 1379845, 2042673, 2566724, 2234425, 1945737, 1976641, 1525251]
})
top_df
```

```python
#create bar chart on the top five months
x_axis = np.arange(0, len(top_df['Months']))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)
plt.title("Most Popular Months To Travel, Nov-2018 - Oct 2019")
plt.xlabel("Months")
plt.ylabel("Number of Tourists")

bar= plt.bar(x_axis, top_df['Tourists'], color = ['green', 'green', 'green', 'green', 'green', 'green', 'blue', 'blue', 'blue', 'blue', 'blue', 'green', 'green'], alpha=0.75, align='center')
plt.xticks(tick_locations, ['Nov', 'Dec', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct'])
```
